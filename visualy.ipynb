```python
import matplotlib.pyplot as plt
import mplfinance as mpl
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data/AAPL.csv",index_col=0,parse_dates=True,skiprows=[1])
data.index.name = "Date"
data = data.dropna()
for col in ["Open", "High", "Low", "Close", "Volume"]:
    data[col] = pd.to_numeric(data[col], errors="coerce")
data = data.dropna()

print(data.shape)
data.head()
```

```python
df_candle = data[["Open", "High", "Low", "Close", "Volume"]].tail(60).copy()
mpl.plot(
    df_candle,
    type="candle",
    style="charles",
    title=f"AAPL Candlestick Chart — Last 60 Days",
    ylabel="Price (USD)",
    ylabel_lower="Shares\nTraded",
    volume=True,
    figsize=(14, 7)
)
```

```python
close_prices = data["Close"]

box_data = [
    close_prices.iloc[i : i + 20].values
    for i in range(len(close_prices) - 20 + 1)
]

plt.figure(figsize=(14, 6))
plt.boxplot(box_data, showfliers=True)
plt.title(f'AAPL — Boxplot of Closing Prices (20-Day Moving 20)')
plt.xlabel('Moving 20 Index')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()
```
